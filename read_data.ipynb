# Mutations 
### Made csv from txt file excel

```python
mutations_filename= "mutations_brca.csv"
```

```python
mutations = pd.read_csv(directory+mutations_filename)
mutations.head()
```

# common mutations greater than 50 individuals

### Common mutations

TP53
RHGB
BRCA1
BRCA2
JAK1
TTN
TNN
SMYD4
NBPF12
MXRA5


```python
com_mutations = list(mutations['Hugo_Symbol'].value_counts()[mutations['Hugo_Symbol'].value_counts() >=50 ].index)
com_mutations[:10]
```

```python
imp_mutations = ["TP53","RHGB","BRCA1","BRCA2","JAK1", "TTN", "TNN", "SMYD4", "NBPF12", "MXRA5"]
```

```python
for mut in imp_mutations:
    if mut in mutations['Hugo_Symbol'].values:
        print(mut,"Yes")
        if mut not in com_mutations:
            print("appending", mut)
            com_mutations.append(mut)
    else:
        print(mut, "NO")
        #imp_mutations.remove(mut)
        
```

```python
#creating yes no columns for mutations
for val in com_mutations:
    mutations[val] = np.where(mutations["Hugo_Symbol"] == val,1,0)
    
```

```python
mutations.columns[130:]
```

```python
sample_group = mutations.groupby("Tumor_Sample_Barcode")
for i in range(len(com_mutations)):
    grouped_yes= pd.Series(sample_group[com_mutations[i]].value_counts().unstack()[1],name=com_mutations[i])
    data = pd.merge(data,grouped_yes, left_on='Sample Identifier', right_on= 'Tumor_Sample_Barcode',how="left")
```

```python
data.shape
```

```python
sample_group = mutations.groupby('Tumor_Sample_Barcode')

sample_group["Hugo_Symbol"].value_counts()
```

```python
mutations['Variant_Classification'].value_counts()
```

```python
sample_group = mutations.groupby('Tumor_Sample_Barcode')

variant_clasification_count = sample_group["Variant_Classification"].value_counts().unstack().add_suffix('_count').reset_index()
```

```python
data = pd.merge(data,variant_clasification_count, left_on='Sample Identifier', right_on= 'Tumor_Sample_Barcode',how="left")
data.shape
```

## creating columns, number of mutations per chromosome

```python
sample_group = mutations.groupby('Tumor_Sample_Barcode')
```

```python
mutation_counts_by_chrom = sample_group["Chromosome"].value_counts().unstack().add_suffix('_mutcount').reset_index()
```

```python
mutation_counts_by_chrom
```

```python
data = pd.merge(data,mutation_counts_by_chrom, left_on='Sample Identifier', right_on= 'Tumor_Sample_Barcode',how="left")
data.shape
```

```python
data.to_csv('brca_merged_data.csv')
```
